```python
import pandas as pd
import numpy as np
df = pd.read_csv('AQI.csv')
```

```python
df.drop('index',axis=1,inplace=True)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3,n_init='auto')
pred = km.fit_predict(df)
```

```python
km.cluster_centers_
```

```python
df2 = pd.DataFrame(km.cluster_centers_,columns=['AQI','PM1','PM2','TEMP','HUMID'])
```

```python
df2
```

```python
s0 = np.sum(df2.iloc[0]) # 0 th cluster has poorest air quality
```

```python
s1 = np.sum(df2.iloc[1]) # 1 th cluster has best air quality
```

```python
s2 = np.sum(df2.iloc[2]) # 2 th cluster has moderate air quality
```

```python
L = [s0,s1,s2]
L = sorted(L)
if L[0] == s0:
  l0 = 'GOOD'
elif L[0] == s1:
  l1 = 'GOOD'
else:
  l2 = 'GOOD'

if L[1] == s0:
  l0 = 'MODERATE'
elif L[1] == s1:
  l1 = 'MODERATE'
else:
  l2 = 'MODERATE'

if L[2] == s0:
  l0 = 'BAD'
elif L[2] == s1:
  l1 = 'BAD'
else:
  l2 = 'BAD'
```

```python
df['label'] = 'NULL'
```

```python
df['label'][pred==0] = l0
df['label'][pred==1] = l1
df['label'][pred==2] = l2
```
